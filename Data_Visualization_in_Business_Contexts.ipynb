#Introduction
Data visualization is the graphical representation of data using charts, graphs, and maps to make complex information more understandable and actionable. In Data Science, it plays a crucial role in:

* Identifying patterns and trends
* Making data-driven decisions
* Communicating insights effectively
* Detecting anomalies and outliers

#Business Benefits of Effective Visualization
* Better Decision-Making – Helps stakeholders understand trends and KPIs quickly.
* Improved Efficiency – Enables faster data analysis and reporting.
* Enhanced Customer Insights – Reveals customer behavior and market trends.
* Competitive Advantage – Allows businesses to identify opportunities and risks.
* Simplified Communication – Converts complex datasets into easy-to-digest visuals.

**Here we use bar chart, the bar chart visualizes the distribution of customers based on Gender. It helps businesses understand the male-to-female ratio among their customers.**
* **Customer Demographics** – Shows whether the customer base is male-dominated, female-dominated, or balanced.
* **Market Segmentation** – Helps identify gender preferences in spending behavior.

```python
age_counts = df["Genre"].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.bar(age_counts.index, age_counts.values, color='skyblue')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

#Age vs. Spending Score Scatter Plot
* **Purpose:** Analyze how customer age influences spending habits.
* **Insight:** Younger customers might have higher spending scores, while older customers could be more cautious.
* **Action:** Develop marketing campaigns tailored to different age groups.


```python
plt.figure(figsize=(10, 5))
plt.scatter(df["Age"], df["Annual_Income_(k$)"], color='blue', alpha=0.6)

# Labels & Title
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.title("Age vs. Annual Income Scatter Plot")
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
gender_counts = df["Genre"].value_counts()

# Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'pink'], startangle=90, shadow=True)

plt.title("Gender Distribution")
plt.show()
```
